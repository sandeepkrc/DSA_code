```python
class Student:
    def __init__(self,a,b):
        self.a = a
        self.b = b

    
    def instance_method(self):
        print("This is instance method")
    
    @staticmethod
    def static_method():
        print("This is static method")

    @classmethod
    def class_method(cls):
        print("this is class method")
```
